## (d) Realice una regresión lineal de mínimos cuadrados básica. Explique la importancia/conveniencia del paso 3 y los argumentos que se deben entregar a la función que implementa la regresión lineal.

```python
X = df_scaled.iloc[:,1:] #use .ix instead, in older pandas version (basicamente elimina el primer parametro "price")
N = X.shape[0]           #Asigna la cantidad de datos en las 17 variables restantes
X.insert(X.shape[1], 'intercept', np.ones(N)) #Asigna el intercepto en 1 a cada valor en la tabla
y = df_scaled['price']                        #Para asi generar la convencion y eliminar B0 y dejar la regresion de forme
#mascara estatica con el 70% de los datos     # Sum(Xgorro*Bj) (Vector Form)
mascara = np.zeros(len(X))
limit = int(len(X)*0.7)
mascara[:limit] = 1 
istrain = mascara== 1                         #Pasos para entrenar
Xtrain = X[istrain]
ytrain = y[istrain]
Xtest = X[np.logical_not(istrain)]
ytest = y[np.logical_not(istrain)]
linreg = lm.LinearRegression(fit_intercept = False)
linreg.fit(Xtrain, ytrain)
bGorro = linreg.coef_
#ver datos de semilla, orden de los datos
```

### Respuesta

Finalmente, se genera la regresion y los valores de esta se guardan en bGorro.

## (e) Construya una tabla con los pesos y Z-score correspondientes a cada predictor (variable). ¿Qué variables están más correlacionadas con la respuesta? Si usáramos un nivel de significación del 5%. ¿Qué es lo que observa y cuál puede ser la causa?

```python
Covariance_Matrix = np.dot(Xtrain.T,Xtrain)
Covariance_Matrix_inventida = np.linalg.inv(Covariance_Matrix)
yGorro = linreg.predict(Xtrain)
ecm = mean_squared_error(ytrain, yGorro)
var_bGorro = ecm*Covariance_Matrix_inventida
X_Columns = list(X.columns)
values = {}
zscore = (bGorro/np.sqrt(var_bGorro.diagonal()))
d = {'pesos': pd.Series(bGorro,index = X_Columns),
     'z-score' : pd.Series(zscore, index = X_Columns)}
final_table = pd.DataFrame(d)
#Las variables mas correlacionadas con la respuesta son en orden decreciente son:
print(final_table.sort_values(by=['z-score'], ascending=[False]))
#Con el 5% de significancia se tiene que los datos con z-score mayor a 1.96 representan los datos mas importantes del
#Conjunto ya que el resto de los datos cae en el 95% restante de la grafica normal, por lo cual, estos datos pudieron
#ser significativos unicamente por el azar
for i in range(18):
    if(abs(final_table['z-score'][i])>1.96):
        print(final_table.index[i])
final_table.head(18)
```

# Selección de Atributos

## (a) Construya una función que implemente Forward Step-wise Selection (FSS). Es decir, partiendo con un modelo sin predictores (variables), agregue un predictor a la vez, re-ajustando el modelo de regresión en cada paso. Para seleccionar localmente una variable, proponga/implemente un criterio distinto al utilizado en el código de ejemplo. Construya un gráfico que muestre el error de entrenamiento y el error de pruebas como función del número de variables en el modelo. Ordene el eje x de menor a mayor.

```python
def fss(x, y, names_x,xPrueba,yPrueba, k = 10000):
    p = x.shape[1]-1
    k = min(p, k)
    ejeX = list(range(0, p))
    names_x = np.array(names_x)
    remaining = list(range(0, p))
    selected = [p]
    current_score = best_new_score = 0.0
    erroresEntrenamiento = []
    erroresPrueba = []
    c=0
    while remaining and len(selected)<=k:
        intercepto = 0
        score_candidates = []
        bGorro = []
        zscore = []
        newZscore = []
        for candidate in remaining:
            model = lm.LinearRegression(fit_intercept=False)
            indexes = selected + [candidate]
            x_train = x[:,indexes]
            yhat_test = model.fit(xPrueba,yPrueba).predict(xPrueba)
            mse_test = mean_squared_error(yhat_test, yPrueba)
            Xm = x_train
            ym = y
            kf = KFold(n_splits=20)
            mse_cv = 0
            error_cv = []
            for train, val in kf.split(Xm):
                linreg = lm.LinearRegression(fit_intercept = False)
                linreg.fit(Xm[train], ym[train])                    
                yhat_val = linreg.predict(Xm[val])                   
                mse_fold = np.mean(np.power(yhat_val - ym[val], 2))
                error_cv.append(mse_fold)
                mse_cv += mse_fold
            mse_cv = mse_cv / 20 ## Cambiar para notar diferencias
            linreg.fit(x_train, y)
            bGorro = linreg.coef_
            Covariance_Matrix = np.dot(x_train.T,x_train)
            Covariance_Matrix_inventida = np.linalg.inv(Covariance_Matrix)
            yGorro = linreg.predict(x_train)
            mse = mean_squared_error(ytrain, yGorro)
            var_bGorro = mse*Covariance_Matrix_inventida
            zsc = bGorro/np.sqrt(var_bGorro.diagonal())
            intercepto = zsc[0]
            zscore.append((zsc[1],mse,mse_cv,candidate))
        best_new_score,best_mse,best_mse_test,best_candidate = zscore.pop()
        remaining.remove(best_candidate)
        selected.append(best_candidate)
        erroresEntrenamiento.append(best_mse)
        erroresPrueba.append(best_mse_test)
    plt.plot(ejeX, erroresEntrenamiento, ".-")
    plt.plot(ejeX, erroresPrueba, ".-") 
    plt.legend(["Entrenamiento", "Prueba"])
    plt.xlabel("Numero de atributos")
    plt.ylabel("Errores")
    plt.show()
    return selected
names_regressors = X.columns[:-1] #without intercept
fss(Xm,ym,names_regressors,Xtest,ytest)




```

## Respuesta (Hace falta hacer zoom para ver el error de entrenamiento)
Como se puede observar, el error de entrenamiento y el error de prueba son practicamente iguales, esto fenomeno ocurre por la poca existencia de predictores (17) ya que podemos estar visualizando la parte mas baja de la "U" la curva de entrenamiento, posiblemente si tubieramos mas variables, veriamos que el error de entrenamiento disminuye y el error de prueba aumenta.

# Regularización

## (a) Ajuste un modelo lineal utilizando “Ridge Regression”, es decir, regularizando con la norma l2. Utilice valores del parámetro de regularización λ ‡ en el rango [10^7 , 10^1 ], variando si estima conveniente. Construya un gráfico que muestre los coeficientes obtenidos como función del parámetro de regularización. Describa lo que observa. (WARNING: Note que la lı́nea 3 y el primer argumento en la lı́nea 9 son crı́ticos).



```python
from sklearn.linear_model import Ridge
import matplotlib.pylab as plt
X2 = X.drop('intercept', axis=1,inplace=False)
Xtrain = X2[istrain]
ytrain = y[istrain]
names_regressors = X2.columns
alphas_ = np.logspace(7,1,base=10)
coefs = []
model = Ridge(fit_intercept=True,solver='svd')
for a in alphas_:
    model.set_params(alpha=a)
    model.fit(Xtrain, ytrain)
    coefs.append(model.coef_)
plt.figure(figsize = (16, 10))
plt.axis([10,10**(7),-0.1,0.2])
ax = plt.gca()
for y_arr, label in zip(np.squeeze(coefs).T, names_regressors):
    plt.plot(alphas_, y_arr, label=label)
plt.legend()

ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1]) # reverse axis
plt.title('Regularization Path RIDGE')
plt.axis('tight')
plt.legend(loc=2,prop={'size':13})
plt.show()
```

## Respuesta
En la linea 3 se elimina el intercepto ya que sera calculado automaticamente por el modelo linear Ridge para hacer las operaciones internas, lo que se observa (leyendo de derecha a izquierda) es que en un inicio el regularizador es muy holgado, pero a medida que nos aproximamos a 10⁷ el regularizador se comporta mas extrictamente haciendo converger los estimadores asintoticamente a 0, por lo cual el modelo tiene un comportamiento tendiendo al underfiting, la maquina tiene a no aprender nuevos modelos y al contrario del overfiting(sobreajuste), no se ajusta a nada.

## (c) Escogiendo uno de los dos métodos regularizadores anteriores, especificando el porqué, construya un gráfico que muestre el error de entrenamiento y el error de pruebas como función del parámetro de regularización. Discuta lo que observa.

```python
%%time
Xtest = X2[np.logical_not(istrain)]
ytest = y[np.logical_not(istrain)]
alphas_ = np.logspace(7,1,base=10)
coefs = []
model = Ridge(fit_intercept=True,solver='svd')
mse_test = []
mse_train = []
k = 100000
for a in alphas_:
    model.set_params(alpha=a)
    model.fit(Xtrain, ytrain)
    yhat_train = model.predict(Xtrain)
    yhat_test = model.predict(Xtest)
    mse_train.append(np.mean(np.power(yhat_train - ytrain, 2)))
    min_mse_train = min(k,np.mean(np.power(yhat_train - ytrain, 2)))
    mse_test.append(np.mean(np.power(yhat_test - ytest, 2)))
    min_mse_test = min(k,np.mean(np.power(yhat_test - ytest, 2)))
plt.figure(figsize = (16, 10))
plt.axis([10,10**(7),0,0.3])
ax = plt.gca()
ax.plot(alphas_,mse_train,label='train error ridge')
ax.plot(alphas_,mse_test,label='test error ridge')
plt.legend(loc=1)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])
plt.show()
print("Minimo Error de Prueba= %s, Minimo Error de Entrenamiento= %s"%(min_mse_test,min_mse_train))
```

```python
%%time
Xtest = X2[np.logical_not(istrain)]
ytest = y[np.logical_not(istrain)]
alphas_ = np.logspace(0,-3,base=10)
coefs = []
model = Lasso(fit_intercept=True)
mse_test = []
mse_train = []
k = 100000
for a in alphas_:
    model.set_params(alpha=a)
    model.fit(Xtrain, ytrain)
    yhat_train = model.predict(Xtrain)
    yhat_test = model.predict(Xtest)
    mse_train.append(np.mean(np.power(yhat_train - ytrain, 2)))
    min_mse_train = min(k,np.mean(np.power(yhat_train - ytrain, 2)))
    mse_test.append(np.mean(np.power(yhat_test - ytest, 2)))
    min_mse_test = min(k,np.mean(np.power(yhat_test - ytest, 2)))
plt.figure(figsize = (16, 10))
plt.axis([10**(-3),1,0,0.3])
ax = plt.gca()
ax.plot(alphas_,mse_train,label='train error ridge')
ax.plot(alphas_,mse_test,label='test error ridge')
plt.legend(loc=1)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])
plt.show()
print("Minimo Error de Prueba= %s, Minimo Error de Entrenamiento= %s"%(min_mse_test,min_mse_train))
```

## Respuesta
Como se puede observar, tanto Bridge como Lasso generar maquinas extremadamente subajustadas, ya que los errores siempre van en aumento, la diferencia es que por implementacion ya que Lasso si transforma los estimadores en 0 por lo cual el error desde cierto punto se mantiene constante, mientras que en Ridge solo se aproxima muy cercanamente a 0 sus estimadores, lo cual genera que el error se aproxime asintoticamente a su valor maximo. Por lo cual mi eleccion seria Ridge, ya que por Ridge tiende a generar mejores resultados cuando se tiene muchos datos y pocas variables mientras que Lasso tiene buen rendimiento cuando se tiene muchas variables. Ademas, como Ridge tiene un mejor tiempo de computo que Lasso y sus errores son practicamente similares, Ridge es la mejor opcion a elegir.
